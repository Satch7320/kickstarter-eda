```python
def make_hist(x, ax, label=None, orientation='vertical', alpha = .7, stacked = False, histtype = 'bar'):
    r = random.random()
    g = random.random()
    b = random.random()
    ax.hist(x, alpha = alpha, color = (r,g,b), orientation = orientation, stacked = stacked, histtype = histtype)
    ax.set(title = label)
    #ax.set_xlabel(x)
    #plt.tight_layout()
    #plt.show()

def subtable(category):
    df = late_csv[late_csv['main_category'] == category]
    return df


def bootstrap(samp, num_bs_sam = 10000):
    bs = []
    
    for _ in range(num_bs_sam):
        bs.append(np.random.choice(samp, size = len(samp)))
        
    return bs

def bootstrap_ci(sample, stat_function=np.mean, num_resamp = 10000, ci = .95):
    bs = bootstrap(sample, num_resamp)
    
    stat_list = []
    
    for row in range(len(bs)):
        stat_list.append(np.percentile(stat_function(bs[row]),ci))
        #print(row)
    
    return np.percentile(sample,ci), stat_list

def series_norm(series):
    return (series - series.mean()) / (series.max() - series.min())

def pledge_per_backers(df, cat, ax):
    r = random.random()
    g = random.random()
    b = random.random()
    
    main = late_csv[late_csv['main_category'] == cat]
    success = main[main['state'] == 'successful']
    new_series = success['pledged']/success['backers']
    
    ax.boxplot(new_series, vert = False, showfliers = False)
```

```python
fv_pledge_sample = f_and_v['pledged'].sample(n=10000,replace=True)
dance = subtable('Dance')
dance_pledge_sample = dance['pledged'].sample(n=10000, replace = True)
fig, ax = plt.subplots()

ax.hist(fv_pledge_sample, label = 'F&V')

ax.hist(dance_pledge_sample, label = 'Dance')

plt.legend()
```

```python
fv_mu = fv_pledge_sample.mean()
fv_dist = stats.norm(loc = fv_mu, scale = fv_pledge_sample.std())

dance_mu = dance_pledge_sample.mean()
dance_dist = stats.norm(loc = dance_mu, scale = dance_pledge_sample.std())

fig, ax = plt.subplots()

x_range = np.linspace(-1000, 70000,num = 10000)

ax.plot(x_range, fv_dist.pdf(x_range), label = 'F&V Pledge Distribution')
ax.plot(x_range, dance_dist.pdf(x_range), label = 'Dance Pledge Distribution')

ax.set(xlim = (-100, 75000), xlabel = 'Pledge amount', ylabel = 'PDF', title = 'Likelihood of Pledge Values')

plt.legend()
plt.savefig('images/pledge_likelihood')

mean1 = fv_mu
std1 = fv_pledge_sample.std()
nobs = 10000

mean2 = dance_mu
std2 = dance_pledge_sample.std()

print(f'H0: There is no difference in revenue based on category popularity.\n')


stat,pval = stats.ttest_ind_from_stats(mean1, std1, nobs, mean2, std2, nobs, equal_var = False)
print(f'With such a small p-value of {pval}, \nwe can reject the null hypothesis and claim that \na more populous category DOES \nhave greater likelihood of larger pledge amounts.')
```
